```python
import pandas as pd
import numpy as np
import os

# ---------- Load raw data ----------
input_path = r"C:\Users\Sakshi Narawade\Documents\BA\Project 3\data\raw\Telecom+Customer+Churn\telecom_customer_churn.csv"
df = pd.read_csv(input_path)

# ---------- Data Cleaning ----------
# 1. Handle Internet-related nulls
internet_cols = [
    'Internet Type', 'Online Security', 'Online Backup',
    'Device Protection Plan', 'Premium Tech Support',
    'Streaming TV', 'Streaming Movies', 'Streaming Music',
    'Unlimited Data'
]
for col in internet_cols:
    df[col] = df[col].fillna('No Internet Service')

# 2. Handle Phone-related nulls
df['Multiple Lines'] = df['Multiple Lines'].fillna('No Phone Service')
df['Avg Monthly Long Distance Charges'] = df['Avg Monthly Long Distance Charges'].fillna(0)

# 3. Handle Offer and Churn Reasons
df['Offer'] = df['Offer'].fillna('No Offer')
df['Churn Category'] = df['Churn Category'].fillna('N/A (Active)')
df['Churn Reason'] = df['Churn Reason'].fillna('N/A (Active)')

# 4. Handle numerical 'GB Download'
df['Avg Monthly GB Download'] = df['Avg Monthly GB Download'].fillna(0)

print("Missing values after cleaning:", df.isnull().sum().sum())

# ---------- Save cleaned dataset ----------
output_path = r"C:\Users\Sakshi Narawade\Documents\BA\Project 3\data\processed\telecom_customer_churn_cleaned.csv"

# Create folder if it doesn't exist
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# Save file
df.to_csv(output_path, index=False)

print("‚úÖ Cleaned dataset successfully saved at:")
print(output_path)

# Optional safety check
print("üìÅ File exists?", os.path.exists(output_path))
```
